```python
import numpy as np

np.random.seed(42)

n_estaciones, n_dias, n_horas, n_vars = 5, 30, 24, 6

datos = np.zeros((n_estaciones, n_dias, n_horas, n_vars))

# Temperatura base por estación + variación diaria + ruido
temp_base = np.array([8, 12, 14, 10, 6]).reshape(5,1,1)
hora = np.arange(24).reshape(1,1,24)
variacion_diaria = 5 * np.sin((hora - 6) * np.pi / 12)
datos[:,:,:,0] = temp_base + variacion_diaria + np.random.normal(0, 2,
(5,30,24))

# Resto de variables (humedad, presión, viento, precipitación, radiación)
datos[:,:,:,1] = np.clip(70 - datos[:,:,:,0] + np.random.normal(0, 10,
(5,30,24)), 30, 100)
datos[:,:,:,2] = 1013 + np.random.normal(0, 5, (5,30,24))
datos[:,:,:,3] = np.abs(np.random.normal(15, 10, (5,30,24)))
datos[:,:,:,4] = np.maximum(0, np.random.exponential(2, (5,30,24)))
radiacion_base = np.maximum(0, 400 * np.sin((hora - 6) * np.pi / 12))
datos[:,:,:,5] = np.clip(radiacion_base + np.random.normal(0, 50, (5,30,24)),
0, 800)

np.save("datos_metereologicos.npy",datos)
```

Ejercicio 11. Condiciones de confort
Define condiciones de confort como: temperatura entre 18°C y 24°C, humedad entre

Ejercicio 12. Normalización de datos
Normaliza cada variable al rango [0, 1] utilizando la fórmula: x_norm = (x - min) / (max -
min). Aplica la normalización a cada variable de forma independiente, usando
broadcasting.

```python
datos_normalizados = np.zeros_like(datos)
variables = ['Temperatura', 'Humedad', 'Presión',
             'Viento', 'Precipitación', 'Radiación']
for i, var in enumerate(variables):
    var_data = datos[:, :, :, i]
    min_val = var_data.min()
    max_val = var_data.max()
    datos_normalizados[:, :, :, i] = (var_data - min_val) / (max_val - min_val)

print(f"  Mínimo normalizado: {datos_normalizados.min()}")
print(f"  Máximo normalizado: {datos_normalizados.max()}")
```

Ejercicio 16. Informe meteorológico automatizado
Crea una función que reciba el array de datos y genere un informe con las siguientes
métricas para cada estación:
• Temperatura: media, mínima, máxima, amplitud térmica media diaria
• Precipitación: total acumulada, número de días con lluvia (>1 mm)
• Viento: velocidad media, porcentaje de horas con viento fuerte (>40 km/h)
• Radiación: energía total acumulada en el mes
El informe debe devolver un array estructurado o un diccionario con los resultados. No
uses bucles: todas las operaciones deben ser vectorizadas.

```python
def generar_informe_meteorologico(datos, estaciones):

    n_est, n_dias, n_horas, _ = datos.shape
    informe = {}

    for i, estacion in enumerate(estaciones):
        temp = datos[i, :, :, 0]
        precip = datos[i, :, :, 4]
        viento = datos[i, :, :, 3]
        radiacion = datos[i, :, :, 5]

        temp_media = temp.mean()
        temp_min = temp.min()
        temp_max = temp.max()

        temp_max_diaria = temp.max(axis=1)
        temp_min_diaria = temp.min(axis=1)
        amplitud_media = (temp_max_diaria - temp_min_diaria).mean()

        precip_total = precip.sum()


        precip_diaria = precip.sum(axis=1)
        dias_lluvia = (precip_diaria > 1).sum()


        viento_medio = viento.mean()


        horas_viento_fuerte = (viento > 40).sum()
        total_horas = viento.size
        pct_viento_fuerte = (horas_viento_fuerte / total_horas) * 100


        energia_total = radiacion.sum()


        informe[estacion] = {
            'temperatura': {
                'media': temp_media,
                'minima': temp_min,
                'maxima': temp_max,
                'amplitud_media_diaria': amplitud_media
            },
            'precipitacion': {
                'total_acumulada': precip_total,
                'dias_con_lluvia': dias_lluvia
            },
            'viento': {
                'velocidad_media': viento_medio,
                'pct_viento_fuerte': pct_viento_fuerte
            },
            'radiacion': {
                'energia_total_mes': energia_total
            }
        }

    return informe
```
